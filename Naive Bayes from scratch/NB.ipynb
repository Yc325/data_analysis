# Preprocessing

```python
#convert to correct label
list_of_rename = {
    'dos':['back.','land.','neptune.','pod.','smurf.','teardrop.'],
    'u2r':['buffer_overflow.','loadmodule.','perl.','rootkit.'],
    'r2l':['ftp_write.','guess_passwd.','imap.','multihop.','phf.','spy.','warezclient.','warezmaster.'],
    'probe':['ipsweep.','nmap.','portsweep.','satan.']}
data['target'] = data['target'].replace({"normal.":"normal"})
for key, item in list_of_rename.items():
    data.loc[data['target'].isin(item), 'target'] = key


#convert cont values into binning 
contValues = ['duration','src_bytes','dst_bytes','wrong_fragment','urgent','hot','num_failed_logins','num_compromised',
'root_shell','su_attempted','num_root','num_file_creations','num_shells','num_access_files','count',
'srv_count','serror_rate','srv_serror_rate','rerror_rate','srv_rerror_rate','same_srv_rate','diff_srv_rate','srv_diff_host_rate',
'dst_host_count','dst_host_srv_count','dst_host_same_srv_rate','dst_host_diff_srv_rate','dst_host_same_src_port_rate',
'dst_host_srv_diff_host_rate','dst_host_serror_rate','dst_host_srv_serror_rate','dst_host_rerror_rate','dst_host_srv_rerror_rate']

for i in contValues:
    data[i] = pd.cut(data[i], bins=4, labels=['low','medium','high','very high'])

#convert boolean columns to Yes and No
bool_columns =['is_guest_login','logged_in','land']
for i in bool_columns:
    data[i] = data[i].replace({0: "no", 1: "yes"})
```

# Naive Bayes Implementation

```python
def stratified_sampling(data):
    """
    Function does stratified spliting
    initially it is group by the target column which is "target"
    
    it splits into 70% and 30%
    """
    train_list = []
    test_list = []
    grouped_df = data.groupby('target')

    # iterate thru each subset of the provided dataset
    for i, g in grouped_df:
        # stratyfing the g class 
        train, test = split(g, [int(.7 * len(g))])

        train_list.append(train)
        test_list.append(test)

    final_train = pd.concat(train_list)
    final_test = pd.concat(test_list)
    
    #to check it it is splitted correctly (check distribution)
    print('test set:')
    print(final_train["target"].value_counts() / len(final_train))
    print(f'Original data:')
    print(data["target"].value_counts() / len(data))
    return final_train,final_test

def count(dataset,colname,label,target):
    """
    function return sum of the values for provided value in a target
    """
    condition = (dataset[colname] == label) & (dataset['target'] == target)
    if colname == 'target':
        return len(dataset[condition])
    #adding 1 to every count to prenevt getting 0s 
    return len(dataset[condition])+1 # + zero counts

def train_model(train):
    """
    following function 
    returns:
    1. Dictionary with probabilities
    2. intials guess for each unique value in target
    3. sum of the each unique values in train dataset
    """
    train_dataset = train.copy()
    #assign dictinary that will get for each unqiue value in target probablity of each column and it is value
    probabilities = {'normal':{},'u2r':{},'dos':{},'r2l':{},'probe':{}}
    
    #caluclate sum of each unqiue value in target
    count_normal = count(train_dataset,'target','normal','normal')
    count_u2r = count(train_dataset,'target','u2r','u2r')
    count_dos = count(train_dataset,'target','dos','dos')
    count_r21 = count(train_dataset,'target','r2l','r2l')
    count_probe = count(train_dataset,'target','probe','probe')
    
    #assign intial guess for each unqiue value in a target
    initial_guess_normal = count_normal/len(train_dataset)
    initial_guess_u2r = count_u2r/len(train_dataset)
    initial_guess_dos = count_dos/len(train_dataset)
    initial_guess_r21 = count_r21/len(train_dataset)
    initial_guess_probe = count_probe/len(train_dataset)

    #loop thru each columns and set their probability to each unique value in a target class
    for col in train_dataset.columns[:-1]:
        probabilities['normal'][col] = {}
        probabilities['u2r'][col] = {}
        probabilities['dos'][col] = {}
        probabilities['r2l'][col] = {}
        probabilities['probe'][col] = {}
        
        #as we added zero counts we need to add all 1ns to the sum of selected label
        add_black_box = len(train_dataset[col].unique())
        for category in train_dataset[col].unique():
            #count values existed in column
            count_ct_normal = count(train_dataset,col,category,'normal')
            count_ct_u2r = count(train_dataset,col,category,'u2r')
            count_ct_dos = count(train_dataset,col,category,'dos')
            count_ct_r2l = count(train_dataset,col,category,'r2l')
            count_ct_probe = count(train_dataset,col,category,'probe')
            
            #caluclate probablity using numerical underflow 
            probabilities['normal'][col][category] = log(count_ct_normal/(count_normal + add_black_box))
            probabilities['u2r'][col][category] = log(count_ct_u2r/ (count_u2r + add_black_box))
            probabilities['dos'][col][category] = log(count_ct_dos/ (count_dos + add_black_box))
            probabilities['r2l'][col][category] = log(count_ct_r2l/ (count_r21 + add_black_box))
            probabilities['probe'][col][category] = log(count_ct_probe/ (count_probe + add_black_box))
    
    counts = {
        "normal":count_normal,
        "u2r":count_normal,
        "dos":count_normal,
        "r2l":count_normal,
        "probe":count_normal
    }
            
    return probabilities,\
    initial_guess_normal,\
    initial_guess_u2r,\
    initial_guess_dos,\
    initial_guess_r21,\
    initial_guess_probe,\
    counts

    

def test_model(test_dataset,
               train,
               probabilities,
               initial_guess_normal,
               initial_guess_u2r,
               initial_guess_dos,
              initial_guess_r21,
              initial_guess_probe,
              counts):
    
    """
    following function calcualtes probablity of each row from the provided dataset
    it takes 
    9 parameters:
    1. Test Dataset
    2. Train dataset (it is used to caluclate probablity of the word if it is not in train dataset (similar to bag of words))
    3. intial guesses
    4. counts
    """
    result = {}
    predicted = []
    #for each row in test dataset we calculate probablity for each unqiue value in target
    #then we set probablity of each value and choose the maximum.
    
    for row in range(0,len(test_dataset)):
        prod_normal = initial_guess_normal
        prod_u2r = initial_guess_u2r
        prod_dos = initial_guess_dos
        prod_r21 = initial_guess_r21
        prod_probe = initial_guess_probe
        
        for feature in data.columns[:-1]:
            #if the provided value not in train dataset 
            #then we caluclate new value probablity by doing same step as was done in train dataset
            #but only for this value.  - > Non zeros smurfing.
            try:
                prod_normal *= probabilities['normal'][feature][test_dataset[feature].iloc[row]]
            except:
                prod_normal *= log(1/counts['normal']+len(train[feature].unique()))
            try:
                prod_u2r *= probabilities['u2r'][feature][test_dataset[feature].iloc[row]]
            except:
                prod_u2r *= log(1/counts['u2r']+len(train[feature].unique()))
                
            try:
                prod_dos *= probabilities['dos'][feature][test_dataset[feature].iloc[row]]
            except:
                prod_dos *= log(1/counts['dos']+len(train[feature].unique())) 
                
            try:
                prod_r21 *= probabilities['r2l'][feature][test_dataset[feature].iloc[row]]
            except:
                prod_r21 *= log(1/counts['r2l']+len(train[feature].unique()))
                
            try:
                prod_probe *= probabilities['probe'][feature][test_dataset[feature].iloc[row]]
            except:
                prod_probe *= log(1/counts['probe']+len(train[feature].unique()))
                

        
        result["normal"]=prod_normal
        result["u2r"] = prod_u2r
        result["dos"]= prod_dos
        result["r2l"]=prod_r21
        result["probe"]=prod_probe
        predicted.append(max(result, key=result.get))
        
    return predicted


def Classification_NB_Accuracy(predicted,data_target):
    """
    following function return accuracy of the prediction
    """
    count = 0
    correct = 0
    incorrect = 0
    total_number_pred = len(predicted)
    accuracy = None
    for i in range(total_number_pred):
        if predicted[i]==data_target['target'].iloc[i]:
            correct+=1
        else:
            incorrect+=1
    accuracy = correct/total_number_pred
    print(f"Accuracy: {accuracy}")
    print(f'Correct: {correct} ')
    print(f'Incorrect {incorrect}')
    return accuracy
```

```python
train, test = stratified_sampling(data)
```

```python
probabilities,prob_normal, prob_u2r, prob_dos, prob_r21, prob_probe, counts = train_model(train)
```

```python
result = test_model(test,train,probabilities,prob_normal,prob_u2r,prob_dos,prob_r21,prob_probe,counts)
```

```python
accuracy = Classification_NB_Accuracy(result,test)
```
